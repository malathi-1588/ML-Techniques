```python
import pandas as pd
import numpy as np
import statistics
```

```python
from sklearn import datasets
mnist = datasets.load_digits()
```

```python
data = pd.DataFrame(mnist.data)
```

```python
cov_mat = data.cov()
```

```python
cov_mat = cov_mat.values
```

```python
cov_mat
```

```python
values, vectors = np.linalg.eig(cov_mat)
```

```python
values
```

```python
vectors
```

```python
PCA = pd.DataFrame()
for i in range(len(vectors)):
    pca = np.dot(data - np.mean(data), vectors[i])
    PCA['PCA'+str(i)] = pca
```

```python
PCA
```

```python
var_frame = pd.DataFrame(PCA.var())
```

```python
var_frame
```

```python
name = var_frame.index
var = var_frame[0]
```

```python
pca_var = list(zip(var, name))
```

```python
pca_var.sort(reverse = True)
```

```python
var, indices = zip(*(pca_var))
```

```python
var
```

```python
indices
```

```python
var = list(var)
sum_var = sum(var)
per_var = []
for i in range(len(var)):
    var[i] = var[i]/sum_var
    per_var.append(var[i])
```

```python
per_var
```

```python
cumvar = np.cumsum(per_var)
```

```python
cumvar = list(cumvar)
for i in range(len(cumvar)):
    cumvar[i] = cumvar[i]*100
    cumvar.append(cumvar[i])
cumvar
```

```python
for i in range(len(cumvar)):
    if (cumvar[i]>90):
        print(i)
        break
```
