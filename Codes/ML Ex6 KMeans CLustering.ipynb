```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
data = pd.read_csv(r'D:/Malathi/SEM_4/ML/data.csv')
```

```python
data.head()
```

```python
dataset = data.iloc[:,[2,3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i, random_state=0)
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters = 6, random_state=0)
y = km.fit_predict(dataset)
y
```

```python
#from mpl_toolkits.mplot3d import Axes3D

sb.set(style = "darkgrid")

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')


ax.scatter(dataset[y==0,0], dataset[y==0,1], dataset[y==0,2], c = 'red')
ax.scatter(dataset[y==1,0], dataset[y==1,1], dataset[y==1,2], c = 'green')
ax.scatter(dataset[y==2,0], dataset[y==2,1], dataset[y==2,2], c = 'yellow')
ax.scatter(dataset[y==3,0], dataset[y==3,1], dataset[y==3,2], c = 'pink')
ax.scatter(dataset[y==4,0], dataset[y==4,1], dataset[y==4,2], c = 'blue')
ax.scatter(dataset[y==5,0], dataset[y==5,1], dataset[y==5,2], c = 'violet')
ax.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], km.cluster_centers_[:,2], s=100, c='black')
plt.show()
```
